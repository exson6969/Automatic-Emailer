```python
import requests 
from bs4 import BeautifulSoup
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
```

```python
li = [] # List of send email 
your_gmail = ''
encrypted_code = ''
```

```python
# BS4 website scraping 
URL = "https://www.netmeds.com/"
r = requests.get(URL)
pro_title = []
pro_final_price = []
pro_mrp = []
img_url = []
pro_href = []

soup = BeautifulSoup(r.content, 'html5lib') 
for i in range(3):
  mydivs = soup.findAll("div", {"class": "pro_detail"})
  mydivs = mydivs[i]
  divs_title = mydivs.find("span", {"class": "cat_title"}).getText()
  divs_final_price = mydivs.find("span", {"id": "final_price"}).getText()
  divs_price = mydivs.find("strike", {"class": "price"}).getText()

  pro_title.append(divs_title)
  pro_final_price.append(divs_final_price)
  pro_mrp.append(divs_price)
  images = mydivs.findAll('img')
  for image in images:
    img_url.append(image['src'])
  for a in mydivs.find_all('a', href=True):
    pro_href.append(a['href'])

message = MIMEMultipart("alternative")
message["Subject"] = "Netmeds Today's DealsðŸ”¥ðŸ”¥"

css_code = """
@import url("https://fonts.googleapis.com/css2?family=Open+Sans:wght@300;400;600;700&display=swap");
body {
  padding: 0;
  margin: 0;
  box-sizing: border-box;
  font-family: "Open Sans", sans-serif;
}

.container {
  width: 100%;
  height: 80vh;
  justify-content: center;
  align-items: center;
  display: flex;
}

.card {
  width: 300px;
  height: 440px;
  border-radius: 5px;
  box-shadow: 0 4px 6px 0 rgba(0, 0, 0, 0.2);
}
.card > *:not(img) {
  padding: 5px 10px;
}


.card-body {
  padding: 5px;
}
.row {
  display: flex;
  justify-content: space-between;
  align-items: center;
  padding: 5px;
}

.card-body p {
  color: #3d3d3d;
  margin-bottom: 20px;
  font-size: 14px;
}

.btn-group {
  display: flex;
}

.btn-group .btn a {
  padding: 5px 15px;
  background-color: #28a745;
  color: #fff;
  border-radius: 3px;
  margin-left: -2px;
}
.btn-group a {
  margin: 0 10px;
  text-decoration: none;
  color: #000;
}

"""
html = f"""\
<!DOCTYPE html>
<html>
<head>
<style>
{css_code}
</style>
	<meta https-equiv="Content-Type" content="text/html; charset=utf-8">
	<meta name="viewport" content="width = 600">
</head>
<body>
<h2>Limited Deal <img src="https://upload.wikimedia.org/wikipedia/commons/3/37/Clock.gif" width = 20px height = 20px><h2>

  <div class="container">
  <div class="card">
    <img src={img_url[0]} />
    <div class="card-body">
      <div class="row">
        <div class="card-title">
          <h4>{pro_title[0]}</h4>
          <h3>{pro_final_price[0]}</h3>
          <h4><s>{pro_mrp[0]}</s></h4>
        </div>
      </div>
      <div class="btn-group">
        <div class="btn">
          <a href="{pro_href[0]}">Buy Now</a>
    </div>
  </div>
</div>
<hr>
  <div class="container">
  <div class="card">
    <img src={img_url[1]}  />
    <div class="card-body">
      <div class="row">
        <div class="card-title">
          <h4>{pro_title[1]}</h4>
          <h3>{pro_final_price[1]}</h3>
          <h4><s>{pro_mrp[1]}</s></h4>
        </div>
      </div>
      <div class="btn-group">
        <div class="btn">
          <a href="{pro_href[1]}">Buy Now</a>
    </div>
  </div>
</div>
<hr>
  <div class="container">
  <div class="card">
    <img src={img_url[2]}  />
    <div class="card-body">
      <div class="row">
        <div class="card-title">
          <h4>{pro_title[2]}</h4>
          <h3>{pro_final_price[2]}</h3>
          <h4><s>{pro_mrp[2]}</s></h4>
        </div>
      </div>
      <div class="btn-group">
        <div class="btn">
          <a href="{pro_href[2]}">Buy Now</a>
    </div>
  </div>
</div>
  
   </body>
   </html>
"""
part2 = MIMEText(html, "html")
message.attach(part2)

for dest in li:
    s = smtplib.SMTP('smtp.gmail.com', 587)
    s.starttls()
    s.login(your_gmail, encrypted_code)

    s.sendmail(your_gmail, dest, message.as_string())
    s.quit()
```
